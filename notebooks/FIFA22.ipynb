```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\pearl\OneDrive\Documents\sports-performance-analysis\sports-performance-analysis\data\FIFA22_official_data.csv")
df.head()
df.shape
df.columns
df.info()
df.describe()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.isnull().sum()
df = df.dropna(subset=['overall','value'])
```

```python
df.sort_values(by='overall', ascending=False)[['name','club','overall']].head(10)
```

```python
sns.scatterplot(data=df, x='overall', y='value')
plt.title("Player Value vs Overall Rating")
plt.show()
```

```python
sns.boxplot(data=df, x='age', y='overall')
plt.title("Age vs Player Performance")
plt.show()
```

```python
df['nationality'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Between Player Attributes")
plt.show()
```
